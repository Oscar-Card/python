Using the data on IMDB movies located at the following url, 

-obtain the 20 highest rated movie directors based on their average movie IMDB ratings.

-Only consider directors with a minimum of 4 movies, otherwise you get misrepresentative data.
Take movies of year >= 1960.

-Print the top 20 highest rated directors with their movies ordered desc on rating

```python
import pandas as pd
dataUrl = 'https://raw.githubusercontent.com/sundeepblue/movie_rating_prediction/master/movie_metadata.csv'
data = pd.read_csv(dataUrl)
data.head()
```

```python
#Year filter
data = data[data['title_year']>=1960]
data.head()
```

```python
#director name count

x= data.director_name.value_counts()
x
```

```python
# filter for director name with a minimum of 4 movies

x[x>=4]
x
```

```python
# Listing movie director name
list(x[x>=4].keys())
```

```python
selected = data.loc[data.director_name.isin(list(x[x>=4].keys()))]
```

```python
selected.groupby('director_name')['imdb_score'].mean().sort_values(ascending=False).head(20)
```
