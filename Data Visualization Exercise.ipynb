**Scatter plots**

https://seaborn.pydata.org/generated/seaborn.scatterplot.html.

Using the following mpg dataset from Seaborn's data repository, create a scatter plot of miles per gallon against against horsepower. In seaborn, the hue parameter determines which column in the data frame should be used for colour encoding. Also, transparency of a plot is managed by the parameter alpha between 0 an 1. Use these parameters in your scatter plot to enrich your display.

```python
import seaborn as sns
sns.set(style="white")

# Load the example mpg dataset
mpg = sns.load_dataset("mpg")
mpg.head()
```

```python
# Answer:

print("mpg Vs Horsepower")
sns.scatterplot(data=mpg, x="mpg", y="horsepower",hue="origin",alpha=0.9)
```

Pair plots display us the distribution of a single variable along with pairwise relationships between 
variables in our data set. 

Using the paiplot function in Seaborn library (https://seaborn.pydata.org/generated/seaborn.pairplot.html), 
explore the following socioeconomic data set obtained from Gapminder (https://www.gapminder.org/data/).

Utilize the hue parameter in your pairplots. What relationships do you identify between pairs of columns?

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/harslan/Data/master/gapminder_data.csv'
data = pd.read_csv(url)
data.head()
```

```python
# Import Seaborn visualization library
import seaborn as sns
sns.set(style="white")

sns.scatterplot(data=data, x="year", y="lifeExp",hue="continent",alpha=1)
plt.show()
```
